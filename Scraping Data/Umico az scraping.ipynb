```python
import requests
from bs4 import BeautifulSoup as bs
import os
import pandas as pd
import numpy as np
```

```python
links = pd.read_csv('umico_az_links.csv')
links = links['All links'].to_list()
```

# ==========================================================
# ==================     Scraping all smartphones      ==================
# ==========================================================

```python
part1 = links[0:800]
part2 = links[800:1600]
part3 = links[1600::]
```

```python
specs = ['Brend:', 'Qurğunun növü:', 'Seriya:', 'Form-faktor:', 'SİM kartının növü:', 'Ekranın diaqonalı, düym:', 'Operativ yaddaşın həcmi:',
         'Ekranın qətnaməsi, Mp:', 'Yaddaşın əsas həcmi:', 'Ekran texnologiyası:', 'Prosessor:', 'Prosessorun nüvələr sayı:',
         'Daxili yaddaş həcmi:', 'Yaddaş kartlarının dəstəklənməsi:',  'Yaddaş kartları üçün yuvalar:', 'Əsas kamera, Mp:', 
         'Əsas kamera dəsti, Mp:', 'Ön kamera, Mp:', 'Wi-Fi standartı:', 'Bluetooth standartı:' ,'Qoşulma portu:', 
         'Simsiz enerji toplama:', 'Korpusun qorunması:', 'Akkumulyatorun həcmi, mA/saat:', 'Rabitə standartları:', 
         'Digər funksiyalar və xüsusiyyətlər:', 'Bazarın başlanğıc ili:', 'İstehsalçı ölkə:', 'Brendin ölkəsi:', 
         'Komplektasiya:', 'Rəng:', 'Korpusun materialı:', 'Hündürlüyü, mm:','Çəki, q:', 'Qalınlığı, mm:', 'Eni, mm:', 'Price']

all_phone_specs = {a:[] for a in specs}


for link in part3:
    request = requests.get(link)
    soup = bs(request.content, 'html.parser')
    
    if link == part3[250]:
        print('30%')
    elif link == part3[500]:
        print('60%')
    elif link == part3[720]:
        print('90%')
    
    try:
        btn = soup.find("a", {"class": "Features-All"})
        btn.get('onclick')
        
        all_specs = soup.find_all('div', {'class':'MPProductDescriptionContentFeatures'})
        all_specs = [x.text  for x in all_specs[1].find_all('td')]

        price = soup.find_all('div', {'class':'MPProductMainDesc-OfferPrice'})[0].text
        all_phone_specs['Price'].append(price)
        
        for spec in specs[:-1]:
            try:
                all_phone_specs[spec].append(all_specs[all_specs.index(spec)+1])


            except:
                all_phone_specs[spec].append(np.nan)
                
        
                
    except:
        continue
```

# ==========================================================

```python
df_part3 = pd.DataFrame(all_phone_specs)
df_part3.to_csv('Umico_part3_data.csv', index=False)
```
